```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\Data Analytics\Datasets\Extarcted\SMS Spam Collection Dataset\spam.csv', encoding = 'latin1')
df = df.iloc[:,:2]
df.columns = ['label', 'message']
df.head()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(df)):
    review = re.sub('^a-zA-Z', ' ', df['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
y = pd.get_dummies(df['label'])
y = y.iloc[:,0].astype(int).values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500, ngram_range = (1,2))
```

```python
len(X_train), len(y_train)
```

```python
X_train_cv = cv.fit_transform(X_train).toarray()
X_test_cv = cv.transform(X_test).toarray()
```

```python
np.set_printoptions(edgeitems = 30, linewidth = 100000,
                   formatter = dict(float = lambda x: "%.3g" % x))
X_train_cv
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train_cv, y_train)
```

```python
y_pred = mnb.predict(X_test_cv)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features = 2500, ngram_range = (1,2))
X_train_tf = tfidf.fit_transform(X_train).toarray()
X_test_tf = tfidf.transform(X_test).toarray()
```

```python
np.set_printoptions(edgeitems = 30, linewidth = 100000,
                   formatter = dict(float = lambda x: "%.3g" % x))
X_train_tf
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train_tf, y_train)
```

```python
y_pred_tf = mnb.predict(X_test_tf)
```

```python
print(accuracy_score(y_test, y_pred_tf))
print(classification_report(y_test, y_pred_tf))
```
